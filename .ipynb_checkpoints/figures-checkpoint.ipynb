### Imports:

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
import matplotlib.ticker as ticker
from matplotlib.patches import Rectangle
import xarray as xr
import cmocean
import seaborn as sns
import pandas as pd
```

### Fig.4:

```python
n_t_fig_data = np.zeros([6,3])
n_ts = [2,6,10,20,30,40]
for i, n_t in enumerate(n_ts):
    data = np.load(f'src/figure data/validation_loss_{n_t}days.npy')
    n_t_fig_data[i,0] = np.mean(data)
    n_t_fig_data[i,1] = np.quantile(data,0.5-0.3413)
    n_t_fig_data[i,2] = np.quantile(data,0.5+0.3413)

n_s_fig_data = np.zeros([8,3])
n_ss = [1000,2000,4000,8000,12000,16000,20000,30000]
for i, n_s in enumerate(n_ss):
    data = np.load(f'src/figure data/validation_loss_30days_{n_s}.npy')
    n_s_fig_data[i,0] = np.mean(data)
    n_s_fig_data[i,1] = np.quantile(data,0.5-0.3413)
    n_s_fig_data[i,2] = np.quantile(data,0.5+0.3413)


fig, axs = plt.subplots(1,2,figsize=(16,8), constrained_layout = True)
axs[0].plot(n_ts, n_t_fig_data[:,0], 'k-', marker = '.', markersize=10)
axs[0].fill_between(n_ts, n_t_fig_data[:,1], n_t_fig_data[:,2], alpha = 0.5)
axs[0].set_ylim([0,0.5])
axs[0].set_xlim([0,41])


axs[1].plot(n_ss, n_s_fig_data[:,0], 'k-', marker = '.', markersize=10)
axs[1].fill_between(n_ss, n_s_fig_data[:,1], n_s_fig_data[:,2], alpha = 0.5)
axs[1].set_ylim([0,0.5])
axs[1].set_xlim([0,31000])
axs[0].set_xlabel('Length of time-series [days]',fontsize=25)
axs[0].set_ylabel(r'$\mathcal{L}$',fontsize=25)
axs[1].set_xlabel('Number of training examples',fontsize=25)
axs[1].set_ylabel(r'$\mathcal{L}$',fontsize=25)
axs[0].tick_params(axis='both', which='major', labelsize=20)
axs[0].tick_params(axis='both', which='minor', labelsize=20)
axs[1].tick_params(axis='both', which='major', labelsize=20)
axs[1].tick_params(axis='both', which='minor', labelsize=20)
```

### Fig. 6:

```python
map_name = 'DUACS'
rmse = np.load('src/figure data/error timeseries/rmse'+map_name+'.npy')
mean_duacs = np.mean(rmse)
dataset = pd.DataFrame({'Method':[map_name],'RMSE [cm]':[100*rmse[0]]})
# dataset.update({'method':'DUACS','ref':'yes','RMSE':rmse_duacs_plot[1]})
for i in range(1,np.size(rmse)):
    dataset = dataset.append({'Method':'DUACS','RMSE [cm]':100*rmse[i]},ignore_index=True)
# for i in range(np.size(rmse_duacs_plot)):
#     dataset = dataset.append({'method':'DUACS','ref':'no','RMSE':rmse_duacs_plot[i]},ignore_index=True)
map_name = 'BFN QG'
rmse = np.load('src/figure data/error timeseries/rmse'+map_name+'.npy')
for i in range(np.size(rmse)):
    dataset = dataset.append({'Method':map_name,'RMSE [cm]':100*rmse[i]},ignore_index=True)    

map_name = 'DYMOST'
rmse = np.load('src/figure data/error timeseries/rmse'+map_name+'.npy')
for i in range(np.size(rmse)):
    dataset = dataset.append({'Method':map_name,'RMSE [cm]':100*rmse[i]},ignore_index=True)

map_name = 'MIOST'
rmse = np.load('src/figure data/error timeseries/rmse'+map_name+'.npy')
for i in range(np.size(rmse)):
    dataset = dataset.append({'Method':map_name,'RMSE [cm]':100*rmse[i]},ignore_index=True)    
    
map_name = 'ConvLSTM SSH'
rmse = np.load('src/figure data/error timeseries/rmse'+map_name+'.npy')
for i in range(np.size(rmse)):
    dataset = dataset.append({'Method':'ConvLSTM\nSSH','RMSE [cm]':100*rmse[i]},ignore_index=True)    

map_name = '4DVarNet'
rmse = np.load('src/figure data/error timeseries/rmse'+map_name+'.npy')
for i in range(np.size(rmse)):
    dataset = dataset.append({'Method':'4DVarNet\nSSH','RMSE [cm]':100*rmse[i]},ignore_index=True)

map_name = 'ConvLSTM SSH-SST'
rmse = np.load('src/figure data/error timeseries/rmse'+map_name+'.npy')
mean_clstm = np.mean(rmse)
for i in range(np.size(rmse)):
    dataset = dataset.append({'Method':'ConvLSTM\nSSH+SST','RMSE [cm]':100*rmse[i]},ignore_index=True)

fig = plt.figure(figsize = (15,10))
ax = plt.axes()
sns.set_style(style={'axes.facecolor': 'white','xtick.bottom':True,'ytick.left':True,'xtick.color': 'black','ytick.color': 'black'}) 
sns.axes_style('ticks')
# sns.set()
# sns.set(font_scale=1)
sns.violinplot(data = dataset, x="Method",y="RMSE [cm]",scale="area",cut=0,inner=None,palette = {'DUACS':'black','DYMOST':'#85754d','MIOST':'#85754d','BFN QG':'#85754d','4DVarNet\nSSH':'#85754d','ConvLSTM\nSSH':'#33006F', 'ConvLSTM\nSSH+SST':'#33006F'},opacity=1)#'MIOST':'#e4e842','ConvLSTM \n SSH':'#2e5b6f','ConvLSTM \n SSH+SST':'#2e5b6f'},opacity=1)
sns.pointplot(data = dataset, x="Method", y="RMSE [cm]", ci="sd",join = False,color='white',capsize = .1)
plt.ylim([0,15])
plt.xlim([-0.5,6.5])
ax.add_patch(Rectangle((-0.5, 0), 7, 3.25, alpha = 0.2, facecolor = 'black'))
plt.annotate('Altimeter noise',xy = [5.6,2.8],fontsize=12)

ax.set_ylabel('RMSE [cm]',fontsize=24)
ax.set_xlabel('', fontsize = 18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)

plt.hlines(3.25,-0.5,6.5, color = 'black',linestyle='--')
# plt.hlines(100*mean_duacs,-0.5,6.5, color = 'red',linestyle='--')
plt.hlines(100*mean_clstm,-0.5,6.5, color = 'red',linestyle='--')
fig.patch.set_facecolor('white')
```

### Fig. 12:

```python
# function to add value labels
def addlabels(x,y,ax_idx,offset):
    y_round = np.round(y,2)
    ylims = axs[ax_idx].get_ylim()
    shift = (ylims[1]-ylims[0])*offset
    for i in range(len(x)):
        axs[ax_idx].text(i, y[i]-shift, y_round[i], ha = 'center',color='white', fontsize=16)

lat0 = 38
om = 2*np.pi/86164.1
f = 2*om*np.sin(np.pi/180*lat0)
g = 9.81
methods = ['DUACS','BFN','DYMOST','MIOST','convlstm_ssh','4dvarNet_2022','convlstm_ssh_sst']
ekes = np.zeros(len(methods))
for i, method in enumerate(methods):
    eke = np.load('src/figure data/eke'+method+'.npy')
    ekes[i] = np.nanmean(eke)
    
enstrophys = np.zeros(len(methods))
for i, method in enumerate(methods):
    enstrophy = np.load('src/figure data/enstrophy'+method+'.npy')
    enstrophys[i] = np.nanmean(enstrophy)

labels = methods.copy()
labels[1] = 'BFN QG'
labels[4] = 'ConvLSTM \nSSH'
labels[5] = '4DVarNet \nSSH'
labels[6] = 'ConvLSTM \nSSH+SST'

fig, axs = plt.subplots(1,2, figsize = (15,6), constrained_layout=True)
axs[0].bar(labels,ekes,color=['black','#85754d','#85754d','#85754d','#33006F','#85754d','#33006F'])
axs[0].set_ylabel(r'Eddy kinetic energy $(m/s)^2$', fontsize = 24)
axs[0].tick_params(axis='both', which='major', labelsize=12)

addlabels(methods,ekes,0,0.05)

axs[1].bar(labels,enstrophys/f**2,color=['black','#85754d','#85754d','#85754d','#33006F','#85754d','#33006F'])
axs[1].set_ylabel(r'Eddy enstrophy $f^2$', fontsize = 24)
axs[1].tick_params(axis='both', which='major', labelsize=12)

addlabels(methods,enstrophys/f**2,1,0.05)
```
